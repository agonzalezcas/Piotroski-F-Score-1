```python
import json
with open('example.json', 'r') as f:
    filing = json.load(f)
filing
```

```python
def build_row(filename, filing):
    ret = {}
    ret['symbol'] = filing['symbol']
    ret['accessNumber'] = filename
    data = filing['data']
    for obj in data['bs']:
        ret['bs-' + obj['concept']] = obj['value']
    for obj in data['cf']:
        ret['bs-' + obj['concept']] = obj['value']
    for obj in data['ic']:
        ret['ic-' + obj['concept']] = obj['value']
    return ret
```

```python
import os

directory = '2019Q2'
data = []

for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open(directory + '/' + filename,'r') as f:
            data.append(build_row(filename, json.load(f)))
        continue
    else:
        continue
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(data)
```

```python
df.to_csv('output.csv')
```
